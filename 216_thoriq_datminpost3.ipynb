```python
# This cell depends on the successful execution of the cell above it, which imports pandas as pd.
df = pd.read_csv('mahasiswa.csv') # Membaca file CSV ke dalam DataFrame menggunakan pandas

df # Menampilkan DataFrame
```

### Memperbaiki penamaan kolom

Kolom perlu diperbaiki penamaannya agar memudahkan kita dalam mempersiapkan data

```python
pemetaan_kolom = {
    'Jumlah SKS' : 'jumlahSKS',
    'Nilai TBQ' : 'nilaiTBQ'
} # Membuat dictionary untuk memetakan nama kolom lama ke nama kolom baru

df.rename(columns=pemetaan_kolom, inplace=True) # Mengganti nama kolom menggunakan dictionary pemetaan_kolom
df.info() # Menampilkan informasi mengenai DataFrame setelah penggantian nama kolom
```

# **SOAL POST 1**
1. Kategorikan data `nilaiTBQ` menjadi:
* 80 >= A
* 70 > B
* 60 > C
* Selain itu D

```python
for baris, kolom in df.iterrows(): # Melakukan iterasi pada setiap baris DataFrame
    if kolom['nilaiTBQ'] >= 80: # Jika nilaiTBQ lebih dari sama dengan 80
        df.loc[baris, 'gradeTBQ'] = 'A' # Grade TBQ adalah 'A'
    elif kolom['nilaiTBQ'] > 70: # Jika nilaiTBQ lebih dari sama 70
        df.loc[baris, 'gradeTBQ'] = 'B' # Grade TBQ adalah 'B'
    elif kolom['nilaiTBQ'] > 60: # Jika nilaiTBQ lebih dari sama 60
        df.loc[baris, 'gradeTBQ'] = 'C' # Grade TBQ adalah 'C'
    else: # Jika nilaiTBQ kurang dari 60
        df.loc[baris, 'gradeTBQ'] = 'D' # Grade TBQ adalah 'D'

df[['NIM', 'nilaiTBQ', 'gradeTBQ']]
```

## **SOAL POST 2**
2. Buatkan kolom baru dengan nama `SisaSKS` dengan cara mengurangkan 144 SKS dengan `jumlahSKS`

```python
for baris, kolom in df.iterrows():
    df.loc[baris, 'sisaSKS'] = 144 - int(kolom['jumlahSKS']) # Menghitung sisa SKS

df
```

# **SOAL POST 3**
3. Kategorikan data `SisaSKS` menjadi:
* Kurang dari 24 masuk kategori `Plus 1`
* Antara 25 dan 48 masuk kategori `Plus 2`
* Lebih dari 49 `Plus 3`

```python
for baris, kolom in df.iterrows():
    if kolom['sisaSKS'] < 24:
        df.loc[baris, 'kategoriSKS'] = 'Plus 1'
    elif kolom['sisaSKS'] >= 25 and kolom['sisaSKS'] <= 48:
        df.loc[baris, 'kategoriSKS'] = 'Plus 2'
    elif kolom['sisaSKS'] >= 49:
        df.loc[baris, 'kategoriSKS'] = 'Plus 3'
    else:
        df.loc[baris, 'kategoriSKS'] = 'Tidak Terkategori'

df[['NIM', 'sisaSKS', 'kategoriSKS']]
```

# **SOAL POST 4**
4. Transformasi data dengan `label encoding` kolom kategori `nilaiTBQ`

### Transformasi Label Encoding

Label Encoding adalah teknik lain untuk mengubah variabel kategorikal menjadi format numerik. Berbeda dengan One-Hot Encoding yang membuat kolom biner baru untuk setiap kategori, Label Encoding menetapkan angka bulat unik untuk setiap kategori.

**Contoh:**

Jika Anda memiliki kolom 'Ukuran' dengan nilai 'Kecil', 'Sedang', 'Besar':

| Ukuran |
|--------|
| Kecil  |
| Sedang |
| Besar  |
| Kecil  |

Setelah Label Encoding, kolom ini akan menjadi:

| Ukuran (Encoded) |
|-----------------|
| 0               |
| 1               |
| 2               |
| 0               |

Meskipun sederhana, Label Encoding dapat menimbulkan masalah bagi algoritma machine learning karena memperkenalkan urutan atau hierarki yang tidak ada dalam data asli (misalnya, algoritma mungkin menganggap 'Besar' (2) lebih penting daripada 'Kecil' (0)). Oleh karena itu, Label Encoding paling cocok digunakan untuk data kategorikal ordinal (di mana urutan memang penting) atau untuk kolom target dalam klasifikasi.

```python
# 1. Buat kolom gradeTBQ berdasarkan nilaiTBQ
for baris, kolom in df.iterrows():
    if kolom['nilaiTBQ'] >= 80:
        df.loc[baris, 'gradeTBQ'] = 'A'
    elif kolom['nilaiTBQ'] > 70:
        df.loc[baris, 'gradeTBQ'] = 'B'
    elif kolom['nilaiTBQ'] > 60:
        df.loc[baris, 'gradeTBQ'] = 'C'
    else:
        df.loc[baris, 'gradeTBQ'] = 'D'

print("Setelah membuat gradeTBQ:")
print(df[['NIM', 'nilaiTBQ', 'gradeTBQ']].head())

# 2. Label Encoding untuk kolom gradeTBQ
# Method 1: Menggunakan map dengan dictionary
grade_mapping = {'A': 0, 'B': 1, 'C': 2, 'D': 3}
df['gradeTBQ_encoded'] = df['gradeTBQ'].map(grade_mapping)

print("\nSetelah Label Encoding:")
print(df[['NIM', 'nilaiTBQ', 'gradeTBQ', 'gradeTBQ_encoded']].head())

# Method 2: Menggunakan scikit-learn LabelEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['gradeTBQ_encoded_sklearn'] = le.fit_transform(df['gradeTBQ'])

print("\nMapping Label Encoding:")
for i, label in enumerate(le.classes_):
    print(f"{label} -> {i}")

# 3. Analisis hasil
print("\nDistribusi Grade TBQ:")
print(df['gradeTBQ'].value_counts().sort_index())

print("\nDistribusi Grade TBQ (Encoded):")
print(df['gradeTBQ_encoded'].value_counts().sort_index())

print("\nSample data lengkap:")
print(df[['NIM', 'nilaiTBQ', 'gradeTBQ', 'gradeTBQ_encoded', 'gradeTBQ_encoded_sklearn']].head(10))

# 4. Cek konsistensi encoding
print("\nKonsistensi Encoding:")
print("Apakah kedua method encoding sama?", (df['gradeTBQ_encoded'] == df['gradeTBQ_encoded_sklearn']).all())
```

# **SOAL POST 5**
5. Transformasi data dengan `one-hot encoding` kolom kategori `sisaSKS`

### Transformasi One-Hot Encoding

One-Hot Encoding adalah teknik yang digunakan untuk mengubah variabel kategorikal menjadi format numerik yang dapat digunakan oleh algoritma machine learning. Dalam One-Hot Encoding, setiap kategori unik dalam kolom kategorikal diubah menjadi kolom biner baru. Setiap kolom biner ini akan berisi nilai 1 jika baris data tersebut termasuk dalam kategori tersebut, dan 0 jika tidak.

**Contoh:**

Jika Anda memiliki kolom 'Warna' dengan nilai 'Merah', 'Biru', 'Hijau':

| Warna |
|-------|
| Merah |
| Biru  |
| Hijau |
| Merah |

Setelah One-Hot Encoding, kolom ini akan menjadi:

| Merah | Biru | Hijau |
|-------|------|-------|
| 1     | 0    | 0     |
| 0     | 1    | 0     |
| 0     | 0    | 1     |
| 1     | 0    | 0     |

Teknik ini penting karena banyak algoritma machine learning tidak dapat bekerja langsung dengan data kategorikal dalam format teks. One-Hot Encoding membantu merepresentasikan data kategorikal dengan cara yang dapat dipahami oleh model.

```python
import pandas as pd

# 1. Hitung sisaSKS
print("1. Menghitung sisaSKS...")
for baris, kolom in df.iterrows():
    df.loc[baris, 'sisaSKS'] = 144 - int(kolom['jumlahSKS'])

print("Setelah menghitung sisaSKS:")
print(df[['NIM', 'jumlahSKS', 'sisaSKS']].head())

# 2. Buat kolom kategoriSKS
print("\n2. Membuat kategoriSKS...")
for baris, kolom in df.iterrows():
    if kolom['sisaSKS'] < 24:
        df.loc[baris, 'kategoriSKS'] = 'Plus 1'
    elif kolom['sisaSKS'] >= 25 and kolom['sisaSKS'] <= 48:
        df.loc[baris, 'kategoriSKS'] = 'Plus 2'
    elif kolom['sisaSKS'] >= 49:
        df.loc[baris, 'kategoriSKS'] = 'Plus 3'
    else:
        df.loc[baris, 'kategoriSKS'] = 'Tidak Terkategori'

print("Setelah membuat kategoriSKS:")
print(df[['NIM', 'sisaSKS', 'kategoriSKS']].head())

# 3. One-Hot Encoding
print("\n3. Melakukan One-Hot Encoding...")
df_encoded = pd.get_dummies(df, columns=['kategoriSKS'], prefix='kategori')

print("\nSetelah One-Hot Encoding:")
print(df_encoded.filter(like='kategori_').head())

print("\nDataFrame Final dengan One-Hot Encoding:")
print(df_encoded.head())

# 4. Analisis hasil
print("\n4. Analisis Hasil:")
print("Distribusi Kategori SKS:")
print(df['kategoriSKS'].value_counts())

print("\nKolom One-Hot Encoding yang terbentuk:")
one_hot_columns = [col for col in df_encoded.columns if 'kategori_' in col]
print(one_hot_columns)

print("\nSample data dengan one-hot encoding:")
sample_cols = ['NIM', 'sisaSKS'] + one_hot_columns
print(df_encoded[sample_cols].head(10))

# 5. Info dataframe final
print("\n5. Info DataFrame Final:")
print(f"Shape: {df_encoded.shape}")
print(f"Kolom: {df_encoded.columns.tolist()}")
```
